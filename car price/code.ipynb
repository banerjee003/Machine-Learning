## Model 

```python
X = df.drop(columns=['price'])
Y = df['price']
```

```python
X
```

```python
X_one_encode = pd.get_dummies(X, columns = ['model', 'transmission', 'fuelType'], drop_first=True)
X_one_encode
```

```python
X_one_encode = X_one_encode.astype(int)
```

```python
X_one_encode
```

```python
X_label = X
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
cols = ['model', 'transmission', 'fuelType']

for i in cols:
    X_label[i] = enc.fit_transform(X_label[i])

X_label
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scalar = StandardScaler()

X_one_encode[numerical_cols] = scalar.fit_transform(X_one_encode[numerical_cols])
```

```python
X_one_encode
```

```python
lcopy = X_label.copy()
```

```python
X_label = lcopy
```

```python
X_label.columns
```

```python
X_label[X_label.columns] = scalar.fit_transform(X_label[X_label.columns])
X_label
```

```python
from sklearn.model_selection import train_test_split   
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encode, Y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n-p-1)
print(f"The r2 test score is {r2} and the adjusted r2 score is {adjusted_r2} for one hot encoding")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_label, Y, test_size=0.33, random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n-p-1)
print(f"The r2 test score is {r2} and the adjusted r2 score is {adjusted_r2} for label encoding")
```
