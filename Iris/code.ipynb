```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
df.isna().sum()
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3)
model_knn.fit(X,y)
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma='auto')
```

```python
model_svm.fit(X_test,y_test)
```

```python
model_svm.score(X_test,y_test)
```

### Using Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier_svm = GridSearchCV((model_svm),{
    'C' : [1,10,20,30],
    'kernel' : ['linear', 'rbf']
},cv=5, return_train_score=False)
```

```python
classifier_svm.fit(X,y)
```

```python
classifier_svm.cv_results_
```

```python
results_svm = pd.DataFrame(classifier_svm.cv_results_)
```

```python
results_svm
```

```python
results_svm[['param_C','param_kernel','mean_test_score']]
```

```python
classifier_knn = GridSearchCV((model_knn),{
    'n_neighbors' : [5,11,15,21],
    'algorithm' : ['auto', 'brute', 'kd_tree']
},cv=5, return_train_score=False)
```

```python
classifier_knn.fit(X,y)
```

```python
results_knn = pd.DataFrame(classifier_knn.cv_results_)
```

```python
results_knn
```

```python
results_knn[['param_algorithm','param_n_neighbors','mean_test_score']]
```
